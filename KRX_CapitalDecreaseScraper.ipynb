```python
import requests
from bs4 import BeautifulSoup
import re
import time

# OTP 생성을 위한 요청 URL
otp_url = 'https://kind.krx.co.kr/common/stockschedule.do'
headers = {
    'Referer': 'https://kind.krx.co.kr/common/stockschedule.do?method=StockScheduleMain&index=11',
    # 필요한 경우 추가 헤더를 추가합니다.
}

# 연도별로 반복
for year in range(2010, 2024):
    # 월별로 반복
    for month in range(1, 13):
        month_str = f'{month:02d}'  # 2자리 숫자 형식으로 변환
        otp_data = {
            'method': 'searchReductionCapital',
            'forward': 'searchreductioncapital',
            'selMonth': month_str,
            'searchCorpName': '',
            'selYear': str(year),
            'showMonth': month_str,
            'periodType': 'M',
            'selDate': f'{year}-{month_str}-01',
            'submitFlagYn': 'N'
        }

        # POST 요청 보내기
        otp_response = requests.post(otp_url, data=otp_data, headers=headers)
        soup = BeautifulSoup(otp_response.text, 'html.parser')
        
        # 'tr' 태그를 찾고 그 내부에서 'td' 태그를 찾아 회사명, 종목코드, 감자 기준일을 추출합니다.
        for tr in soup.find_all('tr'):
            tds = tr.find_all('td')
            if tds and len(tds) >= 3:  # 'td' 태그가 존재하며 최소한 3개 이상의 'td' 태그가 있는지 확인합니다.
                reduction_base_date = tds[0].text.strip()  # 첫 번째 'td'의 텍스트를 감자 기준일로 사용합니다.
                stock_code_search = re.search(r'\b\d{6}\b', tds[1].text)
                if stock_code_search:
                    stock_code = stock_code_search.group()
                    # 회사명을 추출하기 위한 정규 표현식에서 한글 뿐만 아니라 영문자와 숫자도 포함되도록 수정합니다.
                    company_name_search = re.search(r'[가-힣]+', tr.text)
                    if company_name_search:
                        company_name = company_name_search.group()
                        print(f'회사명: {company_name}, 종목코드: {stock_code}, 감자 기준일: {reduction_base_date}')

        # 잠시 대기하는 것이 좋을 수 있습니다 (서버 부하 방지를 위해)
        time.sleep(0.1)  # 필요한 경우 주석을 해제하고 사용
```
